```python
import numpy as np
import scipy as sp
import scipy.optimize as opt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = lambda x: np.cos(x) - x
```

```python
x = np.linspace(-5, 5, 1000)
y = f(x)
plt.plot(x, y)
plt.axhline(0, color='k')
plt.xlim(-5,5)
```

```python
result = opt.bisect(f, -5, 5)
result
```

```python
plt.plot(x,y)
plt.axhline(0, color='k')
plt.scatter([result], [0], c='r', s=100)
plt.xlim(-5,5)
```
